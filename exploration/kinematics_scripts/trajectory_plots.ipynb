```python
import os
import glob
import numpy as np
import scipy.io as sio
# ------ USER PARAMETERS ------
base_folder = "/media/lolly/Bruh/WAYEEGGAL_dataset/WAYEEG_preprocessed" 


patients = [d for d in os.listdir(base_folder)]
patients.sort()
patients = patients[0:1]  # Just process patient 3 for testing
for pf in patients:
    patient_id = int(pf.strip("P"))

    folder_path = os.path.join(base_folder,pf)
    files = glob.glob(os.path.join(folder_path,f"HS_P{patient_id}_trial*.mat"))

    files.sort()
    files = files[0:1]  # Just process the first file for testing

    for path in files:
        
        mat = sio.loadmat(path,squeeze_me=True,struct_as_record=False)
        kin = np.array(mat["trial_kin"])
```

```python
import plotly.graph_objects as go
import numpy as np

n_points = len(kin[0, :])
time = np.linspace(0, 10, n_points)
x = kin[3, :]
y = kin[7, :]
z = kin[11, :]

def create_interactive_plot(start_idx=0, window_size=100):
    end_idx = min(start_idx + window_size, len(time))  # avoid going out of bounds
    
    fig = go.Figure()
    
    # Current window points
    fig.add_trace(go.Scatter3d(
        x=x[start_idx:end_idx],
        y=y[start_idx:end_idx],
        z=z[start_idx:end_idx],
        mode='markers+lines',
        marker=dict(
            size=8,
            color=time[start_idx:end_idx],
            colorscale='Viridis',
            opacity=0.8,
            colorbar=dict(title='Time')
        ),
        line=dict(color='gray', width=3),
        name=f'Points {start_idx}-{end_idx-1}'
    ))
    
    # Full trajectory for context
    fig.add_trace(go.Scatter3d(
        x=x, y=y, z=z,
        mode='lines',
        line=dict(color='lightgray', width=1),
        opacity=0.3,
        name='Full trajectory'
    ))
    
    fig.update_layout(
        title=f'3D Trajectory - Points {start_idx} to {end_idx-1}',
        scene=dict(
            xaxis_title='X',
            yaxis_title='Y',
            zaxis_title='Z'
        ),
        width=1000,
        height=700
    )
    
    return fig

# In Jupyter, just call:
fig = create_interactive_plot(0, 1700)
fig.show()
```
